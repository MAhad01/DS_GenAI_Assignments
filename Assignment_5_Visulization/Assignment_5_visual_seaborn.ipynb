```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset

x=[1,2,3,4,5]
y=[8,7,6,5,4]

sns.relplot(x=x,y=y)
```

```python
# 2 Generate a dataset of random numbers. Visualize the distribution of a numerical variable.

x=np.random.randn(1000)

sns.histplot(x)
```

```python
def get_positive(x):
    pos_x=[]
    for i in x:
        if i>=0:
            pos_x.append(i)
    return pos_x

def get_negative(x):
    neg_x=[]
    for i in x:
        if i<0:
            neg_x.append(i)
    return neg_x
    
```

```python
# 3. Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values.

teams = ['Team A', 'Team B', 'Team C', 'Team D', 'Team E']
scores = np.random.randint(60, 100, size=len(teams))
df = pd.DataFrame({'Team': teams, 'Score': scores})

sns.barplot(x='Team', y='Score', data=df, palette='viridis')
```

```python
# 4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.

categories = ['A', 'B', 'C', 'D']
data = {
    'Category': np.random.choice(categories, 200),
    'Value': np.random.normal(size=200)
}
df = pd.DataFrame(data)


sns.boxplot(x='Category', y='Value', data=df)
```

```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

n_samples = 200

X1 = np.random.normal(0, 1, n_samples)
X2 = 0.5 * X1 + np.random.normal(0, 1, n_samples)  
X3 = 0.3 * X1 + 0.7 * X2 + np.random.normal(0, 1, n_samples)  


df = pd.DataFrame({
    'Feature1': X1,
    'Feature2': X2,
    'Feature3': X3
})

sns.heatmap(df)
```
